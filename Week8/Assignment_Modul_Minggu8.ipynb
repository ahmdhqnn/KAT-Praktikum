No.1 Linear Regression

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# menyiapkan data sample
df = pd.read_csv('winequality-red.csv')

# pilih fitur dan target
X = df['alcohol']
y = df['quality']

# tambahkan konstanta
X_const = sm.add_constant(X)

# bangun model linear
model = sm.OLS(y, X_const).fit()

# tampilkan hasil regresi
print(model.summary())

# visualisasi garis regresi
plt.scatter(X, y, label='Data Aktual', alpha=0.5)
plt.plot(X, model.predict(X_const), color='red', label='Garis Regresi')
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.title('Regresi Linear Single : ')
plt.legend()
plt.grid()
plt.show()
```

No.2 Mutiple Linear Regresi

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# siapkan data sample
X = df.drop('quality', axis=1)
y = df['quality']

# bangun model
model_multi = LinearRegression()
model_multi.fit(X, y)

# prediksi dan evaluasi R-squared
y_pred = model_multi.predict(X)
r_squared = r2_score(y, y_pred)

# tampilkan R-squared
print(" Hasil R-squared Mutiple Linear Regresi :", r_squared)
```
